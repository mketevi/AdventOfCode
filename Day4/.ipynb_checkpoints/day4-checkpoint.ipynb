```python
import re
def concatLines(lines):
    newLines = []
    newLine = ""
    for counter, line in enumerate(lines[:]):
        if line != "\n":
            newLine+=" "
            newLine+=line.split("\n")[0]
        else:
            newLines.append(newLine)
            newLine = ""
    return newLines

def transformPassports(passports):
    transformedPassports = []
    pattern = "(\w{3})(:)(\S*)"
    for passport in passports:
        paper = {}
        elements = re.findall(pattern, passport)
        for element in elements:
            paper[element[0]] = element[2]

        transformedPassports.append(paper)
    return transformedPassports
```
